```python
def elo_graph(df, base_color, user_list):
    df_new = df[df[base_color].str.contains( user_list[0] + '|' +
                                             user_list[1] + '|' +
                                             user_list[2] + '|' +
                                             user_list[3] + '|' +
                                             user_list[4]
                                              ) == True]
    fig = px.line(df_new, x="exact_date", y= base_color + 'Elo' , title= 'Time based ' + base_color +'Elo', color= base_color)
    fig.show()
```

```python
def merge_tournament(x):
    if 'Rated Classical' in x:
        return 'Rated Classical tournament'
    
    elif 'Rated Bullet'in x:
        return 'Rated Bullet tournament'
    
    elif 'Rated Blitz'in x:
        return 'Rated Blitz tournament'
    
    elif 'Rated UltraBullet'in x:
        return 'Rated UltraBullet tournament'
    
    else:
        return x
    
games_metadata_profile['event_groups'] = games_metadata_profile.Event.apply(merge_tournament)
```

### **Analyze Tournaments**

```python
def filter_tournament(df,event_name):
    return df[df.event_groups == event_name]
```

```python
for tournament in games_metadata_profile.event_groups.unique():
    tournament_name = tournament.replace(" ", "_")
    print(tournament_name)
    globals()[str(tournament_name)] = filter_tournament(games_metadata_profile, tournament)
```

```python
def most_common_openning(df):
    print(f"Most Common Opening for {df.event_groups.iloc[0]} :  {df.groupby(by = 'Opening').count()['Event'].nlargest(1).index[0]}" ) 
    print(f"Players have started a game {df.groupby(by = 'Opening').count()['Event'].nlargest(1).values[0]} times for {len(df)}")
```

```python
for tournament in games_metadata_profile.event_groups.unique():
    tournament_name = tournament.replace(" ", "_")
    most_common_openning(globals()[str(tournament_name)])
    print('\n')
```

```python
def highest_lowest_movements(df):
    df_max = df[df.TotalMoves == df['TotalMoves'].max()]
    print(f"Highest movements for {df.event_groups.iloc[0]} :  {df.TotalMoves.max()} | {df_max['White'].iloc[0]} vs  {df_max['Black'].iloc[0]} ") 
    
    df_min = df[df.TotalMoves == df['TotalMoves'].min()]
    print(f"Lowest movements for {df.event_groups.iloc[0]} :  {df.TotalMoves.min()} | {df_min['White'].iloc[0]} vs  {df_min['Black'].iloc[0]} ") 
```

```python
for tournament in games_metadata_profile.event_groups.unique():
    tournament_name = tournament.replace(" ", "_")
    highest_lowest_movements(globals()[str(tournament_name)])
    print('\n')
```

```python
def fastest_slowest_game(df, base_color):
    
    df_max = df[df[base_color +'_playTime_total'] == df[base_color +'_playTime_total'].max()]
    print(f"Slowest players for {base_color} chess pieces for {tournament}:  {df_max[base_color].iloc[0]} | Total Seconds: {df[base_color +'_playTime_total'].max()} | {df_max['White'].iloc[0]} vs  {df_max['Black'].iloc[0]}") 
    
    df_rev = df[ df[base_color +'_playTime_total'] != 0 ]
    df_min = df_rev[df_rev[base_color +'_playTime_total'] == df_rev[base_color +'_playTime_total'].min()]
    print(f"Fastest players for {base_color} chess pieces for {tournament}:  {df_min[base_color].iloc[0]} | Total Seconds:  {df_min[base_color +'_playTime_total'].min()} | {df_min['White'].iloc[0]} vs  {df_min['Black'].iloc[0]}") 
```

```python
for tournament in games_metadata_profile.event_groups.unique():
    for base_color in ['White', 'Black']:
        tournament_name = tournament.replace(" ", "_")
        fastest_slowest_game(globals()[str(tournament_name)], base_color)
        print('\n')
```

### **Analyze Movements of White Chess Pieces**

```python
def filter_color_based_chess_pieces(df, base_color):
    return df[list(df.filter(like= base_color).columns.values) + list(df.columns[-8:].values)]
```

```python
white_piece_df = filter_color_based_chess_pieces(games_metadata_profile, 'White')
white_piece_df.head(2)
```

```python
white_piece_df_rating = white_piece_df[['White', 'WhiteRatingDiff']]
white_piece_df_rating.dropna(inplace = True)
rated_group = white_piece_df_rating .groupby(by='White')['WhiteRatingDiff'].sum().sort_values(ascending = False)

print(f" Top 5 Rated Players: {rated_group.head(5) }")
print('\n')
print(f" Last 5 Rated Players: {rated_group.tail(5) }")
```

### **Analyze Movements of Black Chess Pieces**

```python
black_piece_df = filter_color_based_chess_pieces(games_metadata_profile, 'Black')
black_piece_df.head(2)
```

```python
black_piece_df_rating = black_piece_df[['Black', 'BlackRatingDiff']]
black_piece_df_rating.dropna(inplace = True)
rated_group_bl = black_piece_df_rating .groupby(by='Black')['BlackRatingDiff'].sum().sort_values(ascending = False)

print(f" Top 5 Rated Players: {rated_group_bl.head(5) }")
print('\n')
print(f" Last 5 Rated Players: {rated_group_bl.tail(5) }")
```
